```python
m=" i am asif intezam. i am a student. i am a good boy. i am Asif"
```

```python
from nltk.tokenize import word_tokenize
```

```python
m1= word_tokenize(m)
m1
```

```python
from nltk.collocations import BigramCollocationFinder, TrigramCollocationFinder, ngrams
```

```python
b=BigramCollocationFinder.from_words(m1)
b.ngram_fd
```

```python
b.ngram_fd.keys()
```

```python
t=TrigramCollocationFinder.from_words(m1)
t.ngram_fd
```

```python
t.ngram_fd.keys()
```

```python
n=ngrams(m1,3)
n
```

```python
for i in n:
    print(i)
```
